```python
def predict_batch(ser, X, batch):
    distances = np.sqrt(((X - ser) ** 2).sum(axis=1))

    nearest_indices = distances.nsmallest(20).index

    nearest_batches = batch.loc[nearest_indices]

    batch_counts = nearest_batches.value_counts()

    max_count = batch_counts.max()
    most_frequent_batches = batch_counts[batch_counts == max_count].index

    for batch_value in nearest_batches:
        if batch_value in most_frequent_batches:
            most_frequent_batch = batch_value
            break

    return most_frequent_batch

def read_dataset(path, usecols=None):
    return pd.read_csv(path, sep="\t", index_col=0, usecols=usecols)


def median_scale(data, clip=None, c=1., exclude=None, axis=0):
    """
    Scale using median and mad over any axis (over columns by default).
    Removes the median and scales the data according to the median absolute deviation.

    To calculate Median Absolute Deviation (MAD) - function "mad" from statsmodels.robust.scale is used
    with arguments "c" equals to 1, hence no normalization is performed on MAD
    [please see: https://www.statsmodels.org/stable/generated/statsmodels.robust.scale.mad.html]

    :param data: pd.DataFrame of pd.Series
    :param clip: float, symmetrically clips the scaled data to the value
    :param c: float, coefficient of normalization used in calculation of MAD
    :param exclude: pd.Series, samples to exclude while calculating median and mad
    :param axis: int, default=0, axis to be applied on: if 0, scale over columns, otherwise (if 1) scale over rows

    :return: pd.DataFrame
    """
    from statsmodels.robust.scale import mad

    if exclude is not None:
        data_filtered = data.reindex(data.index & exclude[~exclude].index)
    else:
        data_filtered = data

    median = 1.0 * data_filtered.median(axis=axis)

    if isinstance(data, pd.Series):
        madv = 1.0 * mad(data_filtered.dropna(), c=c)
        c_data = data.sub(median).div(madv)
    else:
        inv_axis = (axis + 1) % 2  # Sub and div are performed by the other axis
        madv = 1.0 * data_filtered.apply(lambda x: mad(x.dropna(), c=c), axis=axis)
        c_data = data.sub(median, axis=inv_axis).div(madv, axis=inv_axis)

    if clip is not None:
        return c_data.clip(-clip, clip)
    return c_data
```

# Test classifier

```python
from sklearn.metrics import log_loss, f1_score, accuracy_score, confusion_matrix
```

```python
def calc_metrics(true, pred, prob=None, verbose=False):
    metrics_d = {}

    if type(prob) is not type(None):
        cross_ent = log_loss(true, prob)
        metrics_d['Cross_Entropy'] = cross_ent
    
    for av in ['weighted', 'micro', 'macro']:
        f1 = f1_score(true, pred, average=av)
        metrics_d[f"F1_{av}"] = f1

    acc = accuracy_score(true, pred)
    metrics_d['Accuracy'] = acc

    if verbose:
        for name, met in metrics_d.items():
            print(f"{name}: {met}")
    return metrics_d
```

```python
def confusion_matrix_2x2(y_true, y_pred):
    unique_classes = np.unique(np.concatenate([y_true, y_pred]))
    total_TP = 0
    total_FN = 0
    total_FP = 0
    total_TN = 0

    for cls in unique_classes:
        true_current = (y_true == cls)
        pred_current = (y_pred == cls)

        cm = confusion_matrix(true_current, pred_current, labels=[True, False])
        
        TP = cm[0, 0]
        FN = cm[0, 1]
        FP = cm[1, 0]
        TN = cm[1, 1]
        
        total_TP += TP
        total_FN += FN
        total_FP += FP
        total_TN += TN

    df_2x2 = pd.DataFrame({
        "Pred Positive": [total_TP, total_FP],
        "Pred Negative": [total_FN, total_TN]
    }, index=["Actual Positive", "Actual Negative"])
    
    return df_2x2
```

```python
with open('model/ovr_knn_calibrated.pickle', 'rb') as f:
    model = pickle.load(f)
```

```python
ind = raw_features[(~raw_features.index.str.contains('TCGA'))&(raw_features.Batch != 'COREAD_META_GSE104645')].index
sample = sc_features.loc[ind].sample(1000,random_state=42).iloc[:,:-1]

true_pred_probas = pd.DataFrame(model.predict_proba(sample), index=sample.index)
true_pred_probas.columns = model.classes_

true_pred_probas.loc[true_pred_probas[true_pred_probas>0.47].isna().all(axis=1), 'Unclassified'] = 1
y_true = true_pred_probas.idxmax(axis=1)
```

## raw scores

```python
from tqdm import tqdm
sample_df = raw_features.loc[sample.index]
ref_df = raw_features.loc[~raw_features.index.isin(sample.index)]
batch = ref_df['Batch']
ref_df = ref_df.drop(columns='Batch')

batch_pred = {}
probas = {}
for name, ser in tqdm(sample_df.iterrows()):
    
    ser = ser[(~ser.index.isin(['Batch']))&(~ser.isna())]
    X_ref = ref_df[ser.index].dropna()

    ident_batch = predict_batch(ser, X_ref, batch)
    batch_pred[name] = ident_batch

    ref_coh = X_ref[batch==ident_batch]
    
    scaled_features = median_scale(pd.concat([ref_coh,ser.to_frame().T]))
    scaled_features = scaled_features.loc[name]
    
    prob = model.predict_proba(scaled_features.to_frame().T)
    probas[name] = list(prob[0])
```

```python
batch_pred = pd.Series(batch_pred)
probas = pd.DataFrame(probas)
probas.index = model.classes_
probas.loc['Unclassified'] = 0
probas.loc['Unclassified',probas.columns[probas[probas>0.47].isna().all()]] = 1
y_pred = probas.idxmax()
```

```python
probas_pred = probas.T
y_true_wo_uncl = y_true[y_true!='Unclassified']
probas_pred_wo_uncl = probas_pred.iloc[:,:-1].loc[y_true_wo_uncl.index]
y_pred_wo_uncl = y_pred.loc[y_true_wo_uncl.index]
batch_true = sample_df.Batch
```

```python
true_l, pred_l, probas_l = y_true, y_pred, probas_pred
name = 'Class prediction with Unclassified'

metrics_d = calc_metrics(true_l, pred_l, probas_l, verbose=True)

sns.heatmap(pd.crosstab(true_l, pred_l), cmap='Blues', annot=True, fmt='.0f', cbar=False)
plt.xlabel('True labels')
plt.xlabel('Predicted labels')
plt.title("{}, \nF1 = {}".format(name, round(metrics_d['F1_weighted'], 3)))
```

```python
true_l, pred_l, probas_l = y_true_wo_uncl, y_pred_wo_uncl, probas_pred_wo_uncl
name = 'Class prediction without Unclassified'

metrics_d = calc_metrics(true_l, pred_l, probas_l, verbose=True)

sns.heatmap(pd.crosstab(true_l, pred_l), cmap='Blues', annot=True, fmt='.0f', cbar=False)
plt.xlabel('True labels')
plt.xlabel('Predicted labels')
plt.title("{}, \nF1 = {}".format(name, round(metrics_d['F1_weighted'], 3)))
```

```python
true_l, pred_l = batch_true, batch_pred
name = 'Batch prediction via distances, k=20'

metrics_d = calc_metrics(true_l, pred_l, None, verbose=True)

x = confusion_matrix_2x2(true_l, pred_l)
x_norm = x.div(x.sum(axis=1), axis=0)

sns.heatmap(x_norm, cmap='Blues', annot=x, fmt='.0f', cbar=False)

plt.xlabel('True labels')
plt.xlabel('Predicted labels')
plt.title("{}, \nF1 = {}".format(name, round(metrics_d['F1_weighted'], 3)))
```
