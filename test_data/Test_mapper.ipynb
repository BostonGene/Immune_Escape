## custom functions for metrics

```python
from sklearn.metrics import log_loss, f1_score, accuracy_score, confusion_matrix
```

```python
def calc_metrics(true, pred, prob=None, verbose=False): # returns classifier metrics dictionary
    metrics_d = {}

    if type(prob) is not type(None):
        cross_ent = log_loss(true, prob)
        metrics_d['Cross_Entropy'] = cross_ent
    
    for av in ['weighted', 'micro', 'macro']:
        f1 = f1_score(true, pred, average=av)
        metrics_d[f"F1_{av}"] = f1

    acc = accuracy_score(true, pred)
    metrics_d['Accuracy'] = acc

    if verbose:
        for name, met in metrics_d.items():
            print(f"{name}: {met}")
    return metrics_d
```

```python
def confusion_matrix_2x2(y_true, y_pred):  # creates 2x2 confusion matrix for lots of classes
    unique_classes = np.unique(np.concatenate([y_true, y_pred]))
    total_TP = 0
    total_FN = 0
    total_FP = 0
    total_TN = 0

    for cls in unique_classes:
        true_current = (y_true == cls)
        pred_current = (y_pred == cls)

        cm = confusion_matrix(true_current, pred_current, labels=[True, False])
        
        TP = cm[0, 0]
        FN = cm[0, 1]
        FP = cm[1, 0]
        TN = cm[1, 1]
        
        total_TP += TP
        total_FN += FN
        total_FP += FP
        total_TN += TN

    df_2x2 = pd.DataFrame({
        "Pred Positive": [total_TP, total_FP],
        "Pred Negative": [total_FN, total_TN]
    }, index=["Actual Positive", "Actual Negative"])
    
    return df_2x2
```

```python
# Progress bar visualization
from tqdm import tqdm
```

```python
from functions.utils import predict_batch, median_scale
```

```python
# Custom functions to calculate predictions

def get_class_predict(probas):
    probas.loc['Unclassified'] = 0
    probas.loc['Unclassified',probas.columns[probas[probas>0.47].isna().all()]] = 1 # Set threshold
    return probas.idxmax()

def get_batch_and_predicitons_for_samples(sample_df, ref_df, batch, k=20):
    batch_pred = {}
    probas = {}
    
    for name, ser in tqdm(sample_df.iterrows()):
        
        ser = ser[(~ser.isna())&(ser.index!='Batch')] # Can't process NaN features; also should delete true batch
        
        X_ref = ref_df[ser.index].dropna() # Same set of features; be consious, classifier only take full set of features!
    
        ident_batch = predict_batch(ser, X_ref, batch, k=k)
        batch_pred[name] = ident_batch
    
        ref_coh = X_ref[batch==ident_batch]
        
        scaled_features = median_scale(pd.concat([ref_coh,ser.to_frame().T])) #concat with identified cohort and scale
        scaled_features = scaled_features.loc[name]
        
        prob = model.predict_proba(scaled_features.to_frame().T)
        probas[name] = list(prob[0])
    
    batch_pred = pd.Series(batch_pred)
    probas = pd.DataFrame(probas)
    probas.index = model.classes_ 
    
    
    class_predict = get_class_predict(probas)

    probas = probas.T

    return batch_pred, probas, class_predict
```

```python
sample_df = raw_features.loc[sample.index] # Same unscaled samples, as in the ground truth data
batch_true = sample_df.Batch

ref_df = raw_features.loc[~raw_features.index.isin(sample.index)] # Rest of metacohort samples are a reference
batch = ref_df['Batch']
ref_df = ref_df.drop(columns='Batch')

result_metrics = {}
result_metrics_wo_uncl = {}
for k in [20,50,100,200, 500]:
    # run prediction
    batch_pred, probas_pred, y_pred = get_batch_and_predicitons_for_samples(sample_df, ref_df, batch, k=k)

    # calculate metrics for predictions with unclassified

    true_l, pred_l, probas_l = y_true, y_pred, probas_pred
    name = f'Class prediction with Unclassified for k={k}, N={N}'
    
    metrics_d = calc_metrics(true_l, pred_l, probas_l, verbose=True)

    #visualize
    sns.heatmap(pd.crosstab(true_l, pred_l), cmap='Blues', annot=True, fmt='.0f', cbar=False)
    plt.xlabel('True labels')
    plt.xlabel('Predicted labels')
    plt.title("{}, \nF1 = {}".format(name, round(metrics_d['F1_weighted'], 3)))
    plt.show()

    #save to dict
    result_metrics[k] = metrics_d
    
    # calculate metrics for predictions without unclassified
    
    probas_pred_wo_uncl = probas_pred.iloc[:,:-1].loc[y_true_wo_uncl.index]
    y_pred_wo_uncl = y_pred.loc[y_true_wo_uncl.index]
    
    true_l, pred_l, probas_l = y_true_wo_uncl, y_pred_wo_uncl, probas_pred_wo_uncl
    name = f'Class prediction without Unclassified for k={k}, N={N}'
    
    metrics_d = calc_metrics(true_l, pred_l, probas_l, verbose=True)
    
    sns.heatmap(pd.crosstab(true_l, pred_l), cmap='Blues', annot=True, fmt='.0f', cbar=False)
    plt.xlabel('True labels')
    plt.xlabel('Predicted labels')
    plt.title("{}, \nF1 = {}".format(name, round(metrics_d['F1_weighted'], 3)))
    plt.show()

    #save to dict
    result_metrics_wo_uncl[k] = metrics_d

result_metrics = pd.DataFrame(result_metrics)
result_metrics_wo_uncl = pd.DataFrame(result_metrics_wo_uncl)
```

```python
result_metrics.iloc[1:].T.plot.line()
plt.xlabel(f'k neighbors')
plt.ylabel('metric')
plt.title(f'Result metrics with Unclassified, N={N}')
```

```python
result_metrics_wo_uncl.iloc[1:].T.plot.line()
plt.xlabel(f'k neighbors')
plt.ylabel('metric')
plt.title(f'Result metrics w/o Unclassified, N={N}')
```

```python
# Load calculated data
probas_exp_k100 = read_dataset(pardir/'test_data/mapper_test/probabilities_for_shortened_expression_k_100.tsv')
probas_exp_k20 = read_dataset(pardir/'test_data/mapper_test/probabilities_for_shortened_expression_k_20.tsv')

y_pred_k100 = get_class_predict(probas_exp_k100)
y_pred_k20 = get_class_predict(probas_exp_k20)\

batch_exp_k100 = read_dataset(pardir/'test_data/mapper_test/batch_pred_for_shortened_expression_k_100.tsv')['0']
batch_exp_k20 = read_dataset(pardir/'test_data/mapper_test/batch_pred_for_shortened_expression_k_20.tsv')['0']
```

### k=20

```python
# define variables
probas_pred = probas_exp_k20.T
y_pred = y_pred_k20
batch_pred = batch_exp_k20
k=20
N=len(y_pred)

#Take same samples from metacohort
ind = sc_features.index[(sc_features.index.isin(probas_pred.index))]
sample = sc_features.loc[ind].iloc[:,:-1]
batch_true = sc_features.loc[ind].iloc[:,-1:].Batch

#Classifier predictions for scaled data is our ground truth
true_pred_probas = pd.DataFrame(model.predict_proba(sample), index=sample.index)
true_pred_probas.columns = model.classes_

#Get unclassified according to paper threshold
true_pred_probas.loc[true_pred_probas[true_pred_probas>0.47].isna().all(axis=1), 'Unclassified'] = 1
y_true = true_pred_probas.idxmax(axis=1)
y_true_wo_uncl = y_true[y_true!='Unclassified']

# For estimation w/o Unclassified
probas_pred_wo_uncl = probas_pred.iloc[:,:-1]
y_pred_wo_uncl = y_pred.reindex(y_true_wo_uncl.index)
```

```python
true_l, pred_l, probas_l = y_true, y_pred, probas_pred

name = f'Class prediction with Unclassified, from expressions, k={k}, N={N}'

metrics_d = calc_metrics(true_l, pred_l, probas_l, verbose=True)

sns.heatmap(pd.crosstab(true_l, pred_l), cmap='Blues', annot=True, fmt='.0f', cbar=False)
plt.xlabel('True labels')
plt.xlabel('Predicted labels')
plt.title("{}, \nF1 = {}".format(name, round(metrics_d['F1_weighted'], 3)))
```

```python
true_l, pred_l, probas_l = to_common_samples((y_true_wo_uncl, y_pred_wo_uncl, probas_pred_wo_uncl))
name = f'Class prediction w/o Unclassified, from expressions, k={k}, N={N}'

metrics_d = calc_metrics(true_l, pred_l, probas_l, verbose=True)

sns.heatmap(pd.crosstab(true_l, pred_l), cmap='Blues', annot=True, fmt='.0f', cbar=False)
plt.xlabel('True labels')
plt.xlabel('Predicted labels')
plt.title("{}, \nF1 = {}".format(name, round(metrics_d['F1_weighted'], 3)))
```

```python
true_l, pred_l = to_common_samples((batch_true, batch_pred))
name = f'Batch prediction via distances, k={k}, N={N}, from expressions'

metrics_d = calc_metrics(true_l, pred_l, None, verbose=True)

x = confusion_matrix_2x2(true_l, pred_l)
x_norm = x.div(x.sum(axis=1), axis=0)

sns.heatmap(x_norm, cmap='Blues', annot=x, fmt='.0f', cbar=False)

plt.xlabel('True labels')
plt.xlabel('Predicted labels')
plt.title("{}, \nF1 = {}".format(name, round(metrics_d['F1_weighted'], 3)))
```
