## Kallisto output check

```python
pseudocounts = read_dataset(cohort_cohort_path_name/'kallisto_pseudocounts.tsv')

pseudocounts.loc[(pseudocounts.Transcript_Type=='protein_coding')&(~pseudocounts.HUGO_Gene.isin(pipeline_genes)),'Transcript_Type'] = 'protein_coding_unused'

pseudocounts.head()
```

```python
# Aggregate
rna_type_agr_data = pseudocounts.groupby('Transcript_Type').sum().T

#Sort by total sum
rna_type_agr_data = rna_type_agr_data.loc[rna_type_agr_data.sum(axis=1).sort_values().index]
```

```python
#Aggregate Ig genes - they are wrongly estimated by Kallisto, note!
rna_type_agr_data['Ig'] = rna_type_agr_data[rna_type_agr_data.columns[rna_type_agr_data.columns.str.contains('IG_')]].sum(axis=1)

#Aggregate TCR genes - they are wrongly estimated by Kallisto, note!
rna_type_agr_data['TCR_genes'] = rna_type_agr_data[rna_type_agr_data.columns[rna_type_agr_data.columns.str.contains('TR_')]].sum(axis=1)

#Aggregate various noncoding RNA genes
other_noncoding = ['3prime_overlapping_ncrna','macro_lncRNA', 'misc_RNA','polymorphic_pseudogene','rRNA','ribozyme', 'sRNA', 'scaRNA','snRNA', 'snoRNA', 'vaultRNA']
rna_type_agr_data['other_noncoding'] = rna_type_agr_data[other_noncoding].sum(axis=1)

#Save only required RNA types
rna_type_agr_data = rna_type_agr_data[rna_types_order]
```

```python
fig,ax = plt.subplots(figsize=(10,7))

bot_bar_plot(rna_type_agr_data, palette=rna_types_palette,ax=ax, order=rna_types_order)

_ = ax.set_xticklabels(ax.get_xticklabels(),rotation=90)

ax.axhline(10e6, c='crimson')
_ = ax.text(1,11e6,'Lower limit for target protein-coding RNAâ†“', c='crimson',bbox=dict(facecolor='white', alpha=0.3))

ax.set_ylabel('Pseudocounts for transcripts, summed by RNA type')
ax.set_xlabel('Samples')

_ = ax.set_title('RNA types distribution by samples')
```

For estimated pseudocounts for reads by samples, all ok

```python
# Get percentages
rna_type_agr_data_div = rna_type_agr_data.div(rna_type_agr_data.sum(axis=1), axis='index')*100

#Sort by protein_coding RNA percentage
rna_type_agr_data_div = rna_type_agr_data_div.sort_values(by='protein_coding')
```

```python
fig,ax = plt.subplots(figsize=(10,7))

bot_bar_plot(rna_type_agr_data_div, palette=rna_types_palette,ax=ax, order=rna_types_order)

_ = ax.set_xticklabels(ax.get_xticklabels(),rotation=90)

ax.axhline(80, c='crimson')
_ = ax.text(1,76,'Lower limit for target protein-coding RNA%â†‘', c='crimson',bbox=dict(facecolor='white', alpha=0.3))

ax.set_ylabel('RNA type read%')
ax.set_xlabel('Samples')

_ = ax.set_title('RNA types percentage distribution by samples')
```
